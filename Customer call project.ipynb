```python
import pandas as pd
```

```python
customer = pd.read_excel('Customer Call List.xlsx')
customer
```

```python
customer= customer.drop_duplicates()
```

```python
customer
```

```python
customer = customer.drop(columns = 'Not_Useful_Column')
```

```python
customer
```

```python
#customer['Last_Name']= customer['Last_Name'].str.lstrip("...")
#customer['Last_Name']= customer['Last_Name'].str.lstrip("/")
#customer['Last_Name']= customer['Last_Name'].str.rstrip("_")
customer['Last_Name']= customer['Last_Name'].str.strip("123._/")
```

```python
customer['Last_Name']
```

```python
customer
```

```python
customer['Phone_Number']= customer['Phone_Number'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

```python
customer
```

```python
customer['Phone_Number'].astype(str)
```

```python
customer['Phone_Number'].info()
```

```python
print(customer['Phone_Number'].apply(type))  # This will show the type of each value
```

```python
customer['Phone_Number'] = customer['Phone_Number'].astype(str).replace(['nan', 'Na'], '')
```

```python
print(customer['Phone_Number'].apply(type))
```

```python
customer['Phone_Number'] = customer['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
customer
```

```python
customer[['Street','State','Zip_code']] = customer['Address'].str.split(',', n=2, expand=True)
customer
```

```python
customer['Paying Customer']= customer['Paying Customer'].str.replace('Yes','Y')
customer['Paying Customer'] = customer['Paying Customer'].str.replace('No','N')
```

```python
customer
```

```python
customer['Do_Not_Contact']= customer['Do_Not_Contact'].str.replace('Yes','Y')
customer['Do_Not_Contact'] = customer['Do_Not_Contact'].str.replace('No','N')
```

```python
customer['Paying Customer'] = customer['Paying Customer'].str.replace('N/a','')
```

```python
customer['Do_Not_Contact'] = customer['Do_Not_Contact'].astype(str).replace('nan','')
```
